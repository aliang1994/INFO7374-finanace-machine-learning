```python
listyear = []
listmonth = []
listday = []

for s in fb['Date'].astype(str):
    listyear.append(s.split("-")[0])
    listmonth.append(s.split("-")[1])
    listday.append(s.split("-")[2])

fb['Year'] = listyear
fb['Month'] = listmonth
fb['Day'] = listday
```

```python
fb.head(5)
```

```python
fb_month_mean = fb.groupby(['Year','Month']).mean()
fb_month_std = fb.groupby(['Year','Month']).std()
```
